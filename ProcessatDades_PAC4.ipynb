# 2. Adquisió i Preparació de les dades

Les dades s’han obtingut en format csv directament de la pàgina web peaceagreements.org utilitzant els filtres que ofereix la mateixa pàgina. S’han fet dues cerques, primerament una amb els acords firmats des del 2000 fins a l’actualitat i l’altra des del 1990 fins al 2000. Els fitxers contenen un total de 266 columnes i 96 i 44 files respectivament, per la qual cosa es realitzarà un concatenat d’aquests dos conjunts per unificar-ho en un de sol. 
Amb les dues adquisicions de dades obtenim un total de 140 línies amb 5 columnes de variables booleanes, 304 numèriques de les quals 249 són enters i 16 de tipus objecte.


```python
# Lectura dels dos fitxers
df96 = pd.read_csv("pax_data_96_agreements_12-11-20.csv")
df44 = pd.read_csv("pax_data_44_agreements_05-12-20.csv")

# Concatenació dels dos dataframes
pax = pd.concat([df96,df44]) 
```

```python
# Visualització de les dades
pax.head()
```

Es pot observar que disposem de diverses variables, algunes de les quals donen una informació més descriptiva de l'acord de pau com els països els quals han firmat l'acord, el tipus d'acord, quin any ha estat firmat, etc. I d'altres tenen com a valor un número amb l'objectiu d'informar d'una categoria. 

El conjunt de dades està format per un total de 140 files i 266 columnes, d’entre els quals 246 són de tipus enter, 5 de tipus float i 15 de tipus object.

En general hi ha un nombre de camps molt alt, per tant s’hauria de reduir la dimensionalitat per tal de poder fer un anàlisi més òptim de les dades i una posterior visualització. Abans, però, a més d’aquestes columnes, en el processat se’n crearan de noves tals com la separació de l’any, el mes i el dia de la columna Dat i el mapeig dels valors 1, 2 i 3 de les variables numèriques relacionades en Corrupció, Terrorisme, Drogues o Crims en una nova variable que directament proporcioni el nom del delicte en qüestió. Per les variables relacionades amb el gènere també es farà un mapeig directament del número al text descriptiu en concret.



```python
# Creem noves columnes
pax['Countries involved'] = pax['Con'].apply(str.split, args = '/') # Els paisos implicats en el conflicte
pax['Year'] = pax['Dat'].apply(lambda date : int(date.split('-')[2])) # Separem l'any de la data
pax['Month'] = pax['Dat'].apply(lambda date : int(date.split('-')[1])) # Separem el mes de la data
pax['Day'] = pax['Dat'].apply(lambda date : int(date.split('-')[0])) # Separem el dia de la data
# Mirem si ha hagut indicis de corrupció, terrorisme, crims o drogues
pax['isCor'] = pax['Cor'] >= 1
pax['isTerr'] = pax['Terr'] >= 1
pax['isCrOcr'] = pax['SsrCrOcr'] >= 1
pax['isDrugs'] = pax['SsrDrugs'] >= 1
pax['isCrime'] = pax['isDrugs'] | pax['isCrOcr'] | pax['isTerr'] | pax['isCor']
```

Per a la visualització de dades s'utilitzaran els primers 18 atributs que presenta el conjunt de dades i es seleccionaran els atributs amb entrades binàries GeMeNu que pot tenir el valor 1 si apareixen expressions de gènere neutre com “dones i homes” o “nens i nenes” o “fills i filles” o un 0 si no apareixen alguna d’aquestes disposicions; i l’atribut GeFa que pot ser 1 si existeix una provisió relacionada amb famílies (més enllà de mencions incidentals) o un 0 si no apareixen. També les noves variables creades en el pas anterior. 

```python
data = pax[['Countries involved', 'Year', 'Month', 'Day', 'Contp', 'PP', 'PPName', 'Reg', 'AgtId', 'Ver', 'Agt', 'Dat', 'Status',
       'Lgt', 'N_characters', 'Agtp', 'Stage', 'StageSub', 'Part', 'ThrdPart',
       'OthAgr','GeMeNu', 'GeFa', 'isCrime']]
```

```python
data["isCrime"] = data["isCrime"].astype(int)
```
